```python
%matplotlib inline
import numpy as np
from IPython.core.display import display, HTML
display(HTML("<style>.container { width:80% !important; }</style>"))
import pandas as pd
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
import seaborn as sns
sns.set_style("darkgrid",{"font.sans-serif":['simhei', 'Arial']})
sns.set_context("notebook", font_scale=2, rc={"lines.linewidth": 2.5})
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['figure.figsize']=(20,7)
def label(ax):
    for p in ax.patches:
        x=p.get_bbox().get_points()[:,0]
        y=p.get_bbox().get_points()[1,1]
        ax.annotate('{:.0f}'.format(y), (x.mean(), y), ha='center', va='bottom')
    return ax
def readChar(recallFile, preFile):
    crecall = pd.read_csv(recallFile, sep="\\s+", names=["word", "type", "length", "pos",  "tf", "pmi", "le", "re", "ans"])
    cpre = pd.read_csv(preFile, sep="\\s+", names=["word", "context", "cseg","wseg", "pos", "boundary","pinying", "win", "nothing","label","ans"]).drop("nothing", axis=1)
    return crecall, cpre
def readWord(recallFile, preFile):
    wrecall = pd.read_csv(recallFile, sep="\\s+", names=["word", "type", "length", "pos",  "tf", "pmi", "le", "re","ans"])
    wpre = pd.read_csv(preFile, sep="\\s+", names=["word", "context", "wseg","length", "pos", "tf", "pmi", "le", "re", "tfPre", "posWin","pinyin", "nothing", "label","ans"]).drop("nothing", axis=1).drop("pinyin", axis=1)
    return wrecall, wpre
def countChart(recall, tag):
    tmp=recall
    no = tmp[~tmp["ans"]][tag].value_counts().sort_index()
    totals = tmp[tag].value_counts().sort_index()
    tmp= pd.concat([no.rename("f"), totals.rename("t")], axis=1).fillna(0)

    yes_rel = [(j-i) for  i,j in zip(tmp["f"], tmp["t"])]
    no_rel = [i for  i,j in zip(tmp["f"], tmp["t"])]

    f, ax = plt.subplots(1, figsize=(20,5))
    N = len(tmp)
    ind = np.arange(N)    # the x locations for the groups
    width = 0.9      # the width of the bars: can also be len(x) sequence
    p1 = ax.bar(ind, no_rel, width, color='#d62728')
    p2 = ax.bar(ind, yes_rel, width,bottom=no_rel)
    ax.set_ylabel('Count')
    ax.set_title('Count in ' + tag)
    plt.xticks(ind, tmp.index)
    ax.legend((p1[0], p2[0]), ('Undetected', 'Detected'))
    
def ratioChart(recall, tag):
    tmp=recall
    no = tmp[~tmp["ans"]][tag].value_counts().sort_index()
    totals = tmp[tag].value_counts().sort_index()
    tmp= pd.concat([no.rename("f"), totals.rename("t")], axis=1).fillna(0)

    yes_rel = [(j-i)/j*100 for  i,j in zip(tmp["f"], tmp["t"])]
    no_rel = [i/j*100  for  i,j in zip(tmp["f"], tmp["t"])]

    f, ax = plt.subplots(1, figsize=(20,5))
    N = len(tmp)
    ind = np.arange(N)    # the x locations for the groups
    width = 0.9      # the width of the bars: can also be len(x) sequence
    p1 = ax.bar(ind, no_rel, width, color='#d62728')
    p2 = ax.bar(ind, yes_rel, width,bottom=no_rel)
    ax.set_ylabel('Percentage')
    ax.set_title('Ratio in ' + tag)
    plt.xticks(ind, tmp.index)
    ax.legend((p1[0], p2[0]), ('Undetected', 'Detected'))

unknowFile = "unknown.info"
dataPath = "../"
newWordFile = "../test/ans/test.txt.new"
dis = pd.read_csv(newWordFile, sep="\\s+", names=["word", "type", "length", "pos", "tf", "pmi", "le", "re"])
dis.head()
```
